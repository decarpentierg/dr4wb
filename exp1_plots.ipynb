```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib qt
```

```python
%matplotlib inline
```

```python
a = np.load("results/exp1_3.npz")
m_values, total_costs, true_cost = a["m_values"], a["total_costs"], a["true_cost"]
```

```python
total_costs.shape
```

```python
costs_means = np.mean(total_costs, axis=1) / true_cost
costs_std = np.std(total_costs, axis=1) / true_cost
yerr = 2 * costs_std / np.sqrt(total_costs.shape[1])
```

```python
plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(13, 8))
plt.errorbar(m_values, costs_means, yerr=yerr)
plt.fill_between(m_values, costs_means - yerr, costs_means + yerr, alpha=0.3)
plt.grid()
plt.title("Experiment 1")
plt.xlabel("Projection dimension m")
plt.ylabel("Ratio of solution quality")
plt.show()
```
